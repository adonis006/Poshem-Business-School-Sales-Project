##Order Statistics:

```python
unique_order_ids = df['Order ID'].nunique()
print(f"Total number of unique Order IDs: {unique_order_ids}")
```

```python
unique_customer_ids = df['Customer ID'].nunique()
print(f"Total number of unique Customer IDs: {unique_customer_ids}")
```

##Geographical Insights:

```python
#Countries in the dataset
unique_countries = df['Country'].unique()
print(f"Countries included in the dataset: {', '.join(unique_countries)}")
```

```python
#Distribution of orders across different regions
region_distribution = df['Region'].value_counts()
print("Distribution of orders across different regions:\n")
print(region_distribution)
```

```python
#Top 5 cities with the highest number of orders
top_cities = df['City'].value_counts().head(5)
print("Top 5 cities with the highest number of orders:\n")
print(top_cities)
```

##Product Categories:

```python
unique_product_categories = df['Category'].unique()
print(f"Unique products category: {unique_product_categories}")
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("Sales by Product Category:\n")
print(category_sales)
```

```python
top_five_selling_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
print("Top 5 Selling Products:\n")
print(top_five_selling_products)
```

## Customer Segmentation:

```python
distribution_of_segment = df['Segment'].value_counts()
print("Distribution of customers across different segments:\n")
print(distribution_of_segment)
```

```python
average_sales_by_segment = df.groupby('Segment')['Sales'].mean().sort_values(ascending=False)
print("Average order value for each customer segment:\n")
print(average_sales_by_segment)
```

##Shipping Insights:

```python
unique_shipping_modes = df['Ship Mode'].nunique()
print(f"No of unique shipping modes:{unique_shipping_modes}")
```

```python
shipping_mode_by_region = df.groupby(['Region', 'Ship Mode']).size().reset_index(name='Count')
print("Shipping Mode Distribution by Region:")
print(shipping_mode_by_region)
```

```python
shipping_mode_pivot = shipping_mode_by_region.pivot(index='Region', columns='Ship Mode', values='Count').fillna(0)

shipping_mode_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Ship Mode Across Regions')
plt.xlabel('Region')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.legend(title='Ship Mode')
plt.tight_layout()
plt.show()
```

##Sales Distribution

```python
overall_distribution_in_sales_data = df['Sales'].sum()
print(f"overall distribution in sales data: {overall_distribution_in_sales_data}")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Sales'])
plt.title('Box Plot of Sales Data')
plt.xlabel('Sales')
plt.show()
```

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Sales'] < lower_bound) | (df['Sales'] > upper_bound)]

print(f"Number of outliers in Sales data: {len(outliers)}")
print("Outliers in Sales data:\n")
display(outliers.head(7))
```

```python
product_categories_distribution = df['Category'].value_counts()
print('Distribution of orders accross different product categories:\n')
print(product_categories_distribution)
```

```python
product_categories_distribution.plot(kind='bar', figsize=(10, 6))
plt.title('Distribution of Orders across different Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
